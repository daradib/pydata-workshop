## About PostgreSQL and PostGIS

PostGIS extends PostgreSQL, a relational database, to work with spatial data, like points, lines, and polygons. Both PostGIS and PostgreSQL are free/libre open source software. PostgreSQL was originally developed at UC Berkeley.

Relational databases store information in tables containing rows and columns. Like Pandas, each column is of a specific data type. Relational databases are commonly used for data entry and web applications.

Advantages of (relational) databases:

 - Flexibility. The [Structured Query Language](https://en.wikipedia.org/wiki/SQL) (SQL) enables complex queries for extracting or inserting data.
 - Performance. Commonly-used columns can be indexed or cached in memory.
 - Normalization. Each record is stored in a single table, but can be referenced multiple times, avoiding duplication and simplifying updates.
 - Consistency. Constraints and data types prevent missing references and invalid values.
 - Concurrent use. Transactions allow multiple users (local or remote) to read and write data simultaneously.
 - Reliability and redundancy. Data can be served by multiple servers.

It is possible to avoid SQL by using an Object Relational Mapper (ORM) that translates between SQL and an object-oriented language like Python, but today we'll be working directly with SQL, which is flexible and powerful.

### A hypothetical relational database

A tax assessor maintaining property records might create tables for cities, parcels, and buildings. In SQL,

    CREATE TABLE cities (
      id serial PRIMARY KEY,
      name varchar(80) NOT NULL
    )

    CREATE TABLE parcels (
      id serial PRIMARY KEY,
      value real,
      assessed date,
      city int references cities(id) NOT NULL,
    )

    CREATE TABLE buildings (
      id serial PRIMARY KEY,
      built date,
      units int,
      parcel int references parcels(id) NOT NULL,
    )

Advantages of this database:

  - Consistency. The database will enforce primary key and not null constraints and foreign key references.
      - primary key constraint: No records with the same ID can exist.
      - serial data type: The next available ID will automatically be used when a record is added.
      - foreign key reference, not null constraint: Each building references a parcel which in turn references a city. Before a building record can be added, its parcel record must be added. And the parcel record cannot be later deleted without cascading the deletion to any buildings that reference it.
  - Fast aggregate operations.
    - Can quickly calculate the up-to-date average assessed value or number of units in each city. Indexing on the primary key constraint makes this fast.
  - Concurrent use. Multiple clerks can view and modify the data simultaneously.
    
Data is often not so cleanly hierarchically. And in practice, it is imported from multiple sources that do not cleanly reference each other.

Luckily, GIS data contains spatial information. A relational database that is spatially-aware, like PostgreSQL with PostGIS, can work with spatial relationships, for example, whether a parcel is geometrically contained within a city or whether a parcel is near a transit station.

### Geometry columns

So far, we worked with non-spatial data. However, each table has a geometry column named `geom`, which represents points, lines, or polygons. This allows us to work with spatial relationships.

PostGIS includes a `geometry_columns` view. A view is a stored query that acts like a table. Each record (row) of this view describes a geometry column in the database.

```python
query("SELECT * FROM geometry_columns;")
```

Most of these columns should be self-expalanatory. The "MULTI" in the type means that multiple geometries can be stored as a collection in each record, e.g., a route can consist of multiple links, each represented as a line.

The SRID is a unique number that identifies the projection. We reprojected the data to a consistent projection for faster spatial operations between tables. We can view more information on our projection by querying the `spatial_ref_sys` table, also included with PostGIS.

```python
query("SELECT * FROM spatial_ref_sys WHERE srid=2768;")
```

European Petroleum Survey Group (EPSG) is an authority that standardizes these SRID numbers. Our projection is called NAD83(HARN) / California zone 3. Zone 3 in the California State Plane coordinate system includes the Bay Area.

### Simple geometry functions

PostgreSQL includes functions like `lower` that act on strings and aggregate functions like `count`, `sum`, and `average` that act on numbers. PostGIS extends PostgreSQL with functions that act on geometry data types.

Let's look at some examples:

 - [`ST_Length`](http://postgis.net/docs/ST_Length.html)
 - [`ST_Area`](http://postgis.net/docs/ST_Area.html)

Our results will be in units of meters as defined by our projection.

The database stores information internally in a custom binary format, but we can easily serialize it in [well-known text](https://en.wikipedia.org/wiki/Well-known_text) (WKT) format using the [`ST_AsText`](http://postgis.net/docs/ST_AsText.html) function. We can also use [`ST_AsEWKT`](http://postgis.net/docs/ST_AsEWKT.html), which includes projection information, or [`ST_AsGeoJSON`](http://postgis.net/docs/ST_AsGeoJSON.html), which outputs [GeoJSON](https://en.wikipedia.org/wiki/GeoJSON).

```python
for function in ["ST_AsText", "ST_AsEWKT", " ST_AsGeoJSON"]:
    print(query("""
        SELECT {}(geom) AS geom
        FROM actransit_stops
        LIMIT 1;
    """.format(function))['geom'][0])
```

Our projection coordinates aren't very readable. Let's try reprojecting with the `ST_Transform` function to WGS 84, which is in units of degrees latitude/longitude.

```python
# First, need to find the SRID for WGS 84.
query("""
    SELECT *
    FROM spatial_ref_sys
    WHERE srtext LIKE 'GEOGCS["WGS 84%%'
""")
```

```python
query("""
    SELECT ST_AsText(ST_Transform(geom,4326)) AS geom
    FROM actransit_stops
    LIMIT 1;
""")['geom'][0]
```

```python
# Query Alameda County area in square miles.
# We assume that county is entirely covered by non-overlapping parcels.
query("""
    SELECT sum(ST_Area(geom)) / 4047 / 640 AS alameda_blocks_area
    FROM alameda_blocks;
""")
```

```python
# Query Alameda County area in square miles.
# Aggregating blocks and cities have similar values.
query("""
    SELECT sum(ST_Area(geom)) / 4047 / 640 AS alameda_cities_area
    FROM alameda_cities;
""")
```

```python
# Query area of each city in square miles.
query("""
    SELECT name,
           ST_Area(geom) / 4047 / 640 AS area
    FROM alameda_cities
    WHERE name <> 'Alameda County';
""")
```

```python
# Query lengths of AC Transit routes longer than 40 mi.
query("""
    SELECT route,
           ST_Length(geom) / 1609 AS actransit_lengths
    FROM actransit_routes
    WHERE ST_Length(geom) > 40 * 1609
    ORDER BY actransit_lengths DESC
""")
```

```python
# Query BART track mileage in miles.
query("""
    SELECT sum(ST_Length(geom)) / 1609 AS bart_length
    FROM bart_routes;
""")
```

### Comparison geometry functions

In a non-spatial database, like one mantained by our hypothetical tax assessor, the Census blocks table would contain a column that identifies each row's city and county. The total population of a city could be determined by aggregating all of the Census blocks matching that city. In a spatial database, this relationship can be done with spatial functions that compare geometries. 

Let's look at some examples:

 - [`ST_Intersection`](http://postgis.net/docs/ST_Intersection.html): whether interiors intersect
 - [`ST_Within`](http://postgis.net/docs/ST_Within.html):  whether fully contained within the other
 - [`ST_Contains`](http://postgis.net/docs/ST_Contains.html): `ST_Within` with reversed argument order
 - [`ST_Distance`](http://postgis.net/docs/ST_Distance.html): shortest distance
 - [`ST_DWithin`](http://postgis.net/docs/ST_DWithin.html): whether within some distance of the other

```python
# From Introduction to PostGIS by Boundless
Image('http://workshops.boundlessgeo.com/postgis-intro/_images/st_intersects.png')
```

```python
# From Introduction to PostGIS by Boundless
Image('http://workshops.boundlessgeo.com/postgis-intro/_images/st_within.png')
```

```python
# Total population of Berkeley, include all blocks that intersect.
query("""
    SELECT sum(pop_total) AS berkeley_pop
    FROM alameda_blocks,
         berkeley_land
    WHERE ST_Intersects(alameda_blocks.geom, berkeley_land.geom);
""")
```

```python
# Total population of Berkeley, exclude blocks that are not fully within.
query("""
    SELECT sum(pop_total) AS berkeley_pop
    FROM alameda_blocks,
         berkeley_land
    WHERE ST_Within(alameda_blocks.geom, berkeley_land.geom);
""")
```

Should we use `ST_Intersects` or `ST_Within`/`ST_Contains` to aggregate the Census blocks in Berkeley? We can inspect the difference by creating a new table containing blocks that intersect with Berkeley's land polygon but are not fully contained within it. We can then view this table in QGIS.

```python
execute("""
    DROP TABLE IF EXISTS partial_blocks;
    CREATE TABLE partial_blocks AS
      SELECT blocks.fips,
             blocks.geom
      FROM alameda_blocks AS blocks,
           berkeley_land  AS land
      WHERE ST_Intersects(blocks.geom, land.geom)
            AND NOT ST_Within(blocks.geom, land.geom);
""")
```

```python
Image('qgis-partial-blocks.png')
```

Looks like `ST_Intersects` included blocks which are not in Berkeley (upper bound) and `ST_Within`/`ST_Contains` excluded blocks which are within Berkeley (lower bound). More about this later.

A naive approach would be to take an average of the two values. A better approach would be to add more logic to our query. For example, we could include all blocks that are mostly (>50%) covered by Berkeley's polygon or proportionally allocate population. A good query would be efficient and begin with functions that are fast to calculate, like `ST_Intersects`, then `ST_Within`, then calculate the proportion of overlap.

```python
# Bad, inefficient query.
#query("""
#    SELECT sum(pop_total) AS berkeley_pop
#    FROM alameda_blocks AS blocks,
#         berkeley_land  AS land
#    WHERE ST_Area(
#            ST_Intersection(blocks.geom, land.geom)
#          ) / ST_Area(blocks.geom) > 0.5;
#""")
```

```python
# Good, efficient query. 100 times faster. Really.
query("""
    SELECT sum(pop_total) AS berkeley_pop
    FROM alameda_blocks AS blocks,
         berkeley_land  AS land
    WHERE ST_Intersects(blocks.geom, land.geom) AND
          (
            ST_Within(blocks.geom, land.geom) OR
            ST_Area(
              ST_Intersection(blocks.geom, land.geom)
            ) / ST_Area(blocks.geom) > 0.5
          );
""")
```

Checking whether geometeries intersect is relatively fast. The database uses a spatial index to quickly eliminate records outside bounding boxes. In the image below, to check which geometries intersect with the star, the database would first look at the bounding box. Thus, it can eliminate the green line before making exact computations. In practice, PostGIS breaks the bounding box into rectangles, sub-rectangles, etc. for finer indexing. `ST_Intersects` is faster than `ST_Within` because it only needs to find a point of intersection, not ensure that all interior points intersect.

```python
# Count AC Transit bus stops in Berkeley.
query("""
    SELECT count(*)
    FROM actransit_stops,
         alameda_cities
    WHERE name = 'Berkeley'
          AND ST_Within(actransit_stops.geom, alameda_cities.geom);
""")
```

```python
# Query distance between bus stops for BART and the station itself.
query("""
    SELECT station,
           stop,
           routes,
           ST_Distance(bart_stations.geom, actransit_stops.geom) AS distance
    FROM bart_stations,
         actransit_stops
    WHERE station = 'DOWNTOWN BERKELEY'
          AND stop LIKE '%%(Berkeley BART%%'
    ORDER BY distance;
"""
)
```

```python
# Query shortest distance between Oakland BART stations and Berkeley.
query("""
    SELECT station,
           ST_Distance(bart.geom, berkeley.geom) / 1609 AS distance
    FROM bart_stations AS bart,
         berkeley_land AS berkeley,
         alameda_cities AS cities
    WHERE ST_Within(bart.geom, cities.geom)
          AND cities.name = 'Oakland'
    ORDER BY distance;
"""
)
```

### Spatial joins

Relational databases are powerful because they make it easy to combine information from multiple tables (join). Spatial databases can use spatial relationships as the join key. We already used spatial joins in a limited form when comparing geometries between multiple tables.

```python
# Which city is each BART station in?
# This is actually a join.
query("""
    SELECT station,
           name AS city
    FROM bart_stations AS bart,
         alameda_cities AS cities
    WHERE ST_Within(bart.geom, cities.geom)
    ORDER BY city
    LIMIT 5;
""")
```

```python
# Which city is each BART station in?
# Explicit (inner) join syntax.
query("""
    SELECT station,
           name AS city
    FROM bart_stations AS bart
    JOIN alameda_cities AS cities
    ON ST_Within(bart.geom, cities.geom)
    ORDER BY city
    LIMIT 5;
""")
```

```python
# Query baseline racial make-up of Alameda County.
query("""
    SELECT 100. * sum(pop_white) / sum(pop_total) AS pct_white,
           100. * sum(pop_black) / sum(pop_total) AS pct_black,
           100. * sum(pop_asian) / sum(pop_total) AS pct_asian,
           sum(pop_total) AS pop_total
    FROM alameda_blocks AS blocks
""")
```

```python
# Query racial make-up of five "diverse" cities in Alameda County.
query("""
    SELECT name,
           100. * sum(pop_white)  / sum(pop_total) AS pct_white,
           100. * sum(pop_black)  / sum(pop_total) AS pct_black,
           100. * sum(pop_native) / sum(pop_total) AS pct_native,
           100. * sum(pop_asian)  / sum(pop_total) AS pct_asian,
           100. * sum(pop_other)  / sum(pop_total) AS pct_other,
           sum(pop_total) AS pop_total
    FROM alameda_blocks AS blocks
    JOIN alameda_cities AS cities
    ON ST_Within(blocks.geom, cities.geom)
    WHERE name <> 'Alameda County'
    GROUP BY name
    ORDER BY pct_white
    LIMIT 5
""")
```

```python
# Query racial make-up within 200 meters of ten BART stations in Alameda.
query("""
    SELECT station,
           100. * sum(pop_white) / sum(pop_total) AS pct_white,
           100. * sum(pop_black) / sum(pop_total) AS pct_black,
           100. * sum(pop_asian) / sum(pop_total) AS pct_asian,
           sum(pop_total) AS pop_total
    FROM alameda_blocks AS blocks
    JOIN bart_stations  AS stations
    ON ST_DWithin(blocks.geom, stations.geom, 200)
    WHERE pop_total > 0
    GROUP BY station
    ORDER BY pct_white
    LIMIT 10;
""")
```

### Geometry Constructing Functions

"Geometry constructing functions" take geometries as inputs and output new shapes. Here are some examples:

 - [`ST_Centroid`](http://postgis.net/docs/ST_Centroid.html): point near center of mass
 - [`ST_Buffer`](http://postgis.net/docs/ST_Buffer.html): extend boundary outward by buffer distance
 - [`ST_Intersection`](http://postgis.net/docs/ST_Intersection.html): geometry region in common
 - [`ST_Union`](http://postgis.net/docs/ST_Union.html): merged geometry
 - [`ST_Collect`](http://postgis.net/docs/ST_Collect.html): aggregates geometries like `ST_Union`, but much faster because it doesn't dissolve intersecting regions

```python
# From Introduction to PostGIS by Boundless
Image('http://workshops.boundlessgeo.com/postgis-intro/_images/centroid.jpg')
```

```python
# Query racial make-up within 100 meters of BART routes in Alameda.
# Use ST_Centroid to speed up calculations.
query("""
    SELECT route,
           100. * sum(pop_white) / sum(pop_total) AS pct_white,
           100. * sum(pop_black) / sum(pop_total) AS pct_black,
           100. * sum(pop_asian) / sum(pop_total) AS pct_asian,
           sum(pop_total) AS pop_total
    FROM alameda_blocks AS blocks
    JOIN bart_routes    AS routes
    ON ST_DWithin(ST_Centroid(blocks.geom), routes.geom, 100)
    GROUP BY route
    ORDER BY pct_white
    LIMIT 10;
""")
```

Geometries were validated using `ST_IsValid` function and reprojected to NAD83(HARN) / California zone 3 (SRID 2768). Attribute and geometry tables were joined into a new table. Columns were renamed or deleted for convenience.
